## Any Guesses?

This game is an easy guessing game where a player is expected to guess a number in the range of 1 to 10. In turn the computer generates a random number in the same range. The code verifies whether or not the number given by the player is a match with the number randomly generated by the computer. If it is a match it will generate a message if not it will generate a message stating how far was the user entered number from the random number, was it too high or too low?. The player may play more than one game. The code keeps a track of the number of games the player has played and displays the same at the end, when the player wishes to play the game no more.

```python
import random 
again = (input("Welcome! Would you like to play - Any Guesses? (Yes/No): ")).lower()

count = 0

while again == 'yes':
    count += 1
    num = int(input("Enter a number: "))
    rannum = random.randint(1,10)

    print("Input Number:  ",num,"\nRandom Number: ",rannum)

    if num == rannum:
        print("It is a match")
    elif num > rannum:
        print("You have guessed to high. Try again next time!")
    elif num < rannum:
        print("You have guessed to low. Try again next time!")
    else:
        print("There was an error. Try again")
        
    again = (input("Do you wish to play again? (Yes/No): ")).lower()

print()
print("Total number of games played: ",count)
```

## Rock-Paper-Scissors

Rock-Paper-Scissors is a handgame played with 2 or more players. For this code we create our own virtual Rock-Paper-Scissors game with a limitation of only two players. To create this game we need to follow certain criteria which are as follows:<br><blockquote>
1. Rock can defeat scissors by breaking the scissor. <br>
2. Scissors can defeat paper by cutting the paper off. <br>
3. Paper can defeat rock by wrapping its self on the rock. <br>
4. If the choice of both the players are the same the match will be a tie. <br>
5. If the either of the choice does not match the keywords rock, paper or scissors the game will terminate.</blockquote>

```python
input1 = (input("Welcome to Rock-Paper-Scissors! Would you wish to play? (Yes/No): ")).lower()

while input1 != 'no':
    play1 = (input("Player 1 enter your choice: ")).lower()
    play2 = (input("Player 2 enter your choice: ")).lower()
    #input1 = 'Yes'

    def game(in1,in2):
        
        if in1 == in2:
            print("Its a tie! ")
        elif in1 in ['rock','scissors'] and in2 in ['rock','scissors']:
            if in1 == 'rock':
                print("Congratulations to Player 1")
            else:
                print("Congratulations to Player 2")
        elif in1 in ['paper','scissors'] and in2 in ['paper','scissors']:
            if in1 == 'scissors':
                print("Congratulations to Player 1")
            else:
                print("Congratulations to Player 2")
        elif in1 in ['paper','rock'] and in2 in ['paper','rock']:
            if in1 == 'paper':
                print("Congratulations to Player 1")
            else:
                print("Congratulations to Player 2")
        else:
            print("Invalid Entry!")

    game(play1,play2)
    input1 = (input("Do you wish to play again? (Yes/No): ")).lower()
```
